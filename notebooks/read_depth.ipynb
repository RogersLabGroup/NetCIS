```python
from pathlib import Path

import pandas as pd
import seaborn.objects as so
from seaborn import axes_style, plotting_context
from IPython.display import display
```

```python
depth = Path("/home/fisch872/mat/projects/Laura-SB-Analysis/2020_SB-output/GRCm38/full-bam-depth-chr.txt")
df2 = pd.read_csv(depth, sep=r"\s+", header=None, names=["file", "chrom", "avg-depth"])
display(df2)
[ print(x, y) for x, y in df2.iloc[df2["avg-depth"].argmax()].items() ]
```

# per chromosome


```python
g = (
    so.Plot(data=df2, x="avg-depth")
    .add(so.Bars(), so.Hist(bins=100))
    .layout(size=(6,4))
    .theme({**axes_style("whitegrid"), **plotting_context("paper")})
    .label(y="count", title="Average Depth Per Chromosome")
)
g.show()

g = (
    so.Plot(data=df2, x="avg-depth")
    .add(so.Bars(), so.Hist(bins=100))
    .layout(size=(6,4))
    .theme({**axes_style("whitegrid"), **plotting_context("paper")})
    .scale(x="symlog")
    .label(y="count", title="Average Depth Per Chromosome")
)
g.show()
```

# per file over chromosomes

```python
df2_mean = pd.DataFrame({"avg-depth": df2.groupby(["file"])["avg-depth"].mean()})
g = (
    so.Plot(data=df2_mean, x="avg-depth")
    .add(so.Bars(), so.Hist(bins=100))
    .layout(size=(6,4))
    .theme({**axes_style("whitegrid"), **plotting_context("paper")})
    .scale(x="symlog")
    .label(y="count", title="Average Depth Per File - Chromosome Mean")
)
g.show()

df2_median = pd.DataFrame({"avg-depth": df2.groupby(["file"])["avg-depth"].median()})
g = (
    so.Plot(data=df2_median, x="avg-depth")
    .add(so.Bars(), so.Hist(bins=100))
    .layout(size=(6,4))
    .theme({**axes_style("whitegrid"), **plotting_context("paper")})
    .scale(x="symlog")
    .label(y="count", title="Average Depth Per File - Chromosome Median")
)
g.show()
```

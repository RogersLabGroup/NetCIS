```python
from pathlib import Path

import pandas as pd
import seaborn.objects as so
from seaborn import axes_style, plotting_context
from IPython.display import display
```

```python
read_depth_dir = Path("/home/fisch872/mat/projects/Laura-SB-Analysis/2020_SB-output/GRCm38/results-read_depth")
```

```python
bam_file = read_depth_dir / "bam-output.txt"
bam_df = pd.read_csv(bam_file, sep="\t")
display(bam_df)
g = (
    so.Plot(data=bam_df, x="#rname", y="meandepth")   #), color="file", edgestyle="file")
    .add(so.Bar(), so.Dodge())
    .layout(size=(10,4))
    .theme({**axes_style("whitegrid")})
    .theme({**plotting_context("paper")})
)
g.show()
```

```python
full_file = read_depth_dir / "read_depth-output.txt"
full_df = pd.read_csv(full_file, sep="\t")
display(full_df)
g = (
    so.Plot(data=full_df, x="#rname", y="meandepth")   #), color="file", edgestyle="file")
    .add(so.Bar(), so.Dodge())
    .layout(size=(10,4))
    .theme({**axes_style("whitegrid")})
    .theme({**plotting_context("paper")})
)
g.show()
```

```python
# mean depth as a factor aka. mean deapth = 3/00 is 3.00x sequencing depth\
# so for when ratio = 1000 = full read depth x divided by cutadapt filtered read depth

tmp_df = pd.DataFrame(bam_df["#rname"])
tmp_df["meandepth-ratio"] = full_df["meandepth"] / bam_df["meandepth"]
g = (
    so.Plot(data=tmp_df, x="#rname", y="meandepth-ratio")   #), color="file", edgestyle="file")
    .add(so.Bar(), so.Dodge())
    .layout(size=(10,4))
    .theme({**axes_style("whitegrid")})
    .theme({**plotting_context("paper")})
)
g.show()
```
